```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
print(plt.style.available)
```

```python
plt.style.use('bmh')
fig, axes = plt.subplots(figsize=(8,8))
```

```python
styles = plt.style.available
n_styles = len(styles)
n_cols=7
n_rows=4
count = 0

fig = plt.figure(figsize=(18,30))
for style in styles:
  count+=1
  plt.style.use(style)
  plt.subplot(n_cols,n_rows,count)
  plt.plot(x,y)
  plt.title(style)
```

```python
#Example
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'b-')
ax.plot(x,x+3,'g.-')
ax.plot(x,x+4,'y:')
```

## Pyplot

```python
#Example
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,color='green')
ax.plot(x,x+2,color='blue')
ax.plot(x,x+3,color='#66FF66')
ax.plot(x,x+4,color='#CCFFCC')
```

```python
#Transparence with aplha
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,color='green', alpha=1.0)
ax.plot(x,x+2,color='green', alpha=0.7)
ax.plot(x,x+3,color='green', alpha=0.4)
ax.plot(x,x+4,color='green', alpha=0.2)
```

```python
#Linewidth
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,color='green', linewidth=10)
ax.plot(x,x+2,color='green', linewidth=7)
ax.plot(x,x+3,color='green', linewidth=4)
ax.plot(x,x+4,color='green', linewidth=2)
```

Linestyles are `'-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'`

```python
#Line styles
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,color='green', linestyle='-')
ax.plot(x,x+2,color='green', linestyle='dashdot')
ax.plot(x,x+3,color='green', linestyle='dashed')
ax.plot(x,x+4,color='green', linestyle=':')
```

```python
#Markers
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,color='green', marker='8',markersize=10, markerfacecolor='orange')
ax.plot(x,x+2,color='green', marker='v',markersize=12, markerfacecolor='red')
ax.plot(x,x+3,color='green', marker='P',markersize=14, markerfacecolor='yellow')
ax.plot(x,x+4,color='green', marker='o',markersize=16, markerfacecolor='blue')
```
