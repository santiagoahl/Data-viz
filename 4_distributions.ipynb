# Plot distributions in seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#sns.set(rc={"figure.dpi":100}) #Pixels density
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.histplot(data=tips, x='tip', bins = 15, cumulative =True)
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False)
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='frequency')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='probability')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='density')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='density', multiple = 'stack')
plt.show()
```

```python
#Very useful
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='density', multiple = 'dodge')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='density', multiple = 'fill')
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex')
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=True)
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, shade=True)
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, shade=True, bw_adjust=1)
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex', stat='count')
plt.show()
```

```python
sns.displot(data=tips, x='tip', hue='sex', kind='hist', multiple = 'stack')
plt.show()
```

## Displot with other plots

```python
sns.displot(data=tips, x='tip', hue='sex', kind='hist', multiple = 'stack')
plt.show()
```
