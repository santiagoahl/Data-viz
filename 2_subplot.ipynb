# Subplots 

What if I want more than one vizualization?

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

## Usual plot

```python
n_rows= 1
n_cols = 2
#Usual plot
plt.subplot(n_rows, n_cols, 1) #1 means the index
plt.plot(x,y, 'rD-')
#Histogram
plt.subplot(n_rows, n_cols, 2)
plt.hist(y)
plt.show()
```

## Histogram

```python
n_rows= 1
n_cols = 2
#Usual plot
plt.subplot(n_rows, n_cols, 1) #1 means the index
plt.plot(x,y, 'rD-')
plt.plot(y, x, 'rD-')

#Histogram
plt.subplot(n_rows, n_cols, 2)
plt.pie(y)
plt.show()
```

```python
n_rows= 2
n_cols = 2
#Usual plot
plt.subplot(n_rows, n_cols, 1)
plt.plot(x,y, 'rD-')
plt.plot(y, x, 'rD-')
#Pie diagram
plt.subplot(n_rows, n_cols, 2)
plt.pie(y)
#Histogram
plt.subplot(n_rows, n_cols, 3)
plt.hist(y)
#Another histogram
plt.subplot(n_rows, n_cols, 4)
plt.hist(x)
plt.show()
```
