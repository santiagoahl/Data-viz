# How to plot categorical data?

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.countplot(data=tips, x = 'day', hue='sex')
plt.show()
```

```python
sns.stripplot(data=tips, x = 'day', y='total_bill',hue='sex')
plt.show()
```

```python
sns.stripplot(data=tips, x = 'day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.stripplot(data=tips, x = 'day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.swarmplot(data=tips, x = 'day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(data=tips, x = 'day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.swarmplot(data=tips, x = 'day', y='total_bill', hue='sex', dodge=True, color='black', marker = '<')
sns.boxplot(data=tips, x = 'day',y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.violinplot(data=tips, y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.violinplot(data=tips, x = 'day', y='total_bill', hue='sex', palette='pastel', split='True')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.violinplot(data=tips, y='total_bill', hue='sex', split='True')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.catplot(data=tips, x='day',y='total_bill', hue='sex', dodge=True, kind='swarm')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.catplot(data=tips, x='day',y='total_bill', hue='sex', dodge=True, kind='violin')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.catplot(data=tips, x='day',y='total_bill', hue='sex', dodge=True, kind='box', col='time')
plt.show()
```
