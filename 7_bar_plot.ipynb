# Bar plots

The let us plot categorical variables.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA','JAPAN','MEXICO','COLOMBIA','GERMANY']
population = [1000, 800, 900,1000,300]
```

```python
plt.bar(country, population)
plt.show()
```

```python
plt.bar(country, population, width = 0.3, color = ['red', 'blue'])
plt.xticks(np.arange(5),('India', 'Japon', 'Mexico', 'Colombia', 'Alemania'), rotation = 45)
plt.title('Countries population')
plt.show()
```

```python
plt.barh(country,population)
plt.show()
```
