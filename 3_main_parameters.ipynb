# Which are the most used parameters in Seaborn?

They are

- `load_dataset`
- `data`
- `hue`
- `kind`
- `legend`
- `palette`

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set(rc={"figure.dpi":300}) #Pixels density
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
plt.figure(figsize=(20,20))
sns.displot(data=tips) #Distribution plot
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip') #Distribution plot
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip', hue='sex') #Distribution plot
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex') #Distribution plot
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend=False, palette='dark', alpha=0.4) #Distribution plot
plt.show()
```
