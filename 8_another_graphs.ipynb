# How to make other graphics?

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,1000)
```

# Histogram

```python
plt.figure(figsize = (20,5))
plt.hist(data)
plt.show()
```

```python
plt.figure(figsize = (20,5))
plt.hist(data, bins = 50, histtype = 'bar')
plt.show()
```

```python
plt.figure(figsize = (20,5))
plt.hist(data, bins = 50, histtype = 'step')
plt.show()
```

# Box plot

```python
plt.figure(figsize = (10,10))
plt.boxplot(data)
plt.show()
```

```python
plt.figure(figsize = (15,5))
plt.boxplot(data, vert = False, patch_artist=True,notch=True)
plt.show()
```

```python
data = np.append(data, 200)

plt.figure(figsize = (15,5))
plt.boxplot(data, vert = False, patch_artist=True,notch=True)
plt.show()
```

```python
data = np.append(data, 200)

plt.figure(figsize = (15,5))
plt.boxplot(data, vert = False, patch_artist=True,notch=True,showfliers=False)
plt.show()
```

# Scatter (Dispersion graphic)

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = 200*np.random.rand(N)**2
colors = np.random.rand(N)
```

```python
plt.figure(figsize = (10,10))
plt.scatter(x,y)
plt.show()
```

```python
plt.figure(figsize = (10,10))
plt.scatter(x,y, s = area)
plt.show()
```

```python
plt.figure(figsize = (10,10))
plt.scatter(x,y, s = area, c = colors)
plt.show()
```

```python
plt.figure(figsize = (10,10))
plt.scatter(x,y, s = area, c = colors, marker = 'o')
plt.show()
```

```python
plt.figure(figsize = (10,10))
plt.scatter(x,y, s = area, c = colors, marker = 'o',alpha=0.5)
plt.show()
```

# Combine Scatter and Histograms

Thanks to the GridSpec library, a 2-variable scatter plot can be combined with the histograms of each of the 2 variables.
With this library we define a figure of 4x4 graphics. Then we share join zones of the graph grid defined so that:
For the scatter plot there is a 3x3 zone, for the histogram of x a zone of 1x3 and for the histogram of y a zone of 3x1.

For more info see: [Scatter plot with histograms](https://stackabuse.com/matplotlib-scatter-plot-with-distribution-plots-histograms-jointplot/)

```python
import pandas as pd
from matplotlib.gridspec import GridSpec
```

```python
fig = plt.figure()
plt.figure(figsize= (10,10))
gs = GridSpec(4, 4)
ax_scatter = fig.add_subplot(gs[1:4, 0:3])
ax_hist_y = fig.add_subplot(gs[0,0:3])
ax_hist_x = fig.add_subplot(gs[1:4, 3])

plt.show()
```

```python
fig = plt.figure()
plt.figure(figsize= (10,10))
gs = GridSpec(4, 4)
ax_scatter = fig.add_subplot(gs[1:4, 0:3])
ax_hist_y = fig.add_subplot(gs[0,0:3])
ax_hist_x = fig.add_subplot(gs[1:4, 3])

N = 50 
x = np.random.rand(N)
y = np.random.rand(N)

df = pd.DataFrame({'x': x, 'y': y})

ax_scatter.scatter(df.x,df.y)
ax_hist_x.hist(df.x, orientation = 'horizontal')
ax_hist_y.hist(df.y)
plt.show()
```
