```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
#Basic
fig = plt.figure() #Canva
axes = fig.add_axes([0.1,0.1,0.5,0.9]) #Independent figures
axes.plot(x,y,'b')
fig.show()
```

```python
#Adventages
fig = plt.figure() #Canva
axes = fig.add_axes([0.01,0.01,0.9,0.9]) #First two:position, second two: dimensions
axes2 = fig.add_axes([0.3,0.3,0.4,0.3])
axes.plot(x,y,'b')
axes2.plot(y, x,'r')
fig.show()
```

```python
fig = plt.figure() #Canva
axes = fig.add_axes([0.01,0.01,0.9,0.9]) #First two:position, second two: dimensions
axes2 = fig.add_axes([0.3,0.3,0.4,0.3])
axes.plot(x,y,'b')
axes2.set_facecolor('gray') #Not allowed
axes2.plot(y, x,'r')
axes2.set_facecolor('green')
fig.show()
```
