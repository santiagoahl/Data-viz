# Combine several graphs

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Jointplot

```python
sns.jointplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='kde')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True)
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', 
              kind='hist', marginal_ticks=True, marginal_kws=dict(bins=50), multiple='dodge')
plt.show()
```

## Pairplot
 Relation between numeric variables
 

```python
sns.pairplot(data=tips)
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex')
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm', corner=True)
plt.show()
```
