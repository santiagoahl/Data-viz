# What is pyplot

Main `matplotlib` tool to drop graphics.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
#Basic plot
plt.plot(x,y)
plt.show() #Standard show
```

```python
#Color plot
#plt.plot(x,y,'bo')
#plt.plot(x,y,'r')
#plt.plot(x,y,'rx')
plt.plot(x,y,'rs-')
plt.show() 
```

```python
plt.plot(x,y,'yD:')
plt.show() 
```

## Histograms

```python
plt.hist(x,y)
plt.show() 
```

# Pie diagrams

```python
plt.pie(x)
plt.show() 
```

## Scatter

```python
plt.scatter(x,y)
plt.show() 
```

## Box plot

```python
plt.boxplot(x)
plt.show() 
```
