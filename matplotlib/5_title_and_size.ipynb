# Titles, labels, legends and size

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
# Set title
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title('X and Y relation')
axes[1].plot(y,x)
axes[1].set_title('Y and X relation')
```

```python
# Axes labels
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title('X and Y relation')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

axes[1].plot(y,x)
axes[1].set_title('Y and X relation')
axes[1].set_xlabel('Time')
axes[1].set_ylabel('Value')
```

```python
#Fig size
fig, axes = plt.subplots(1,2, figsize=(15,8))
axes[0].plot(x,y)
axes[0].set_title('X and Y relation')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

axes[1].plot(y,x)
axes[1].set_title('Y and X relation')
axes[1].set_xlabel('Time')
axes[1].set_ylabel('Value')
```

```python
#Graph lable
fig, axes = plt.subplots(1,2, figsize=(15,8))
axes[0].plot(x,y, label="$sin(x)$") #Latex notation
axes[0].set_title('X and Y relation')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()

axes[1].plot(y,x)
axes[1].set_title('Y and X relation')
axes[1].set_xlabel('Time')
axes[1].set_ylabel('Value')
```

```python
fig, axes = plt.subplots(1,2, figsize=(15,8))
axes[0].plot(x,y)
axes[0].set_title('X and Y relation')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

axes[1].plot(y,x)
axes[1].set_title('Y and X relation')
axes[1].set_xlabel('Time')
axes[1].set_ylabel('Value')
```
